```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data
accel_data = pd.read_csv('accelerometer_data.csv', names=['Timestamp', 'X', 'Y', 'Z'])

# Convert Timestamp to datetime
accel_data['Timestamp'] = pd.to_datetime(accel_data['Timestamp'], format='%Y%m%d%H%M%S')

# Plot Accelerometer data
plt.figure(figsize=(10, 6))
plt.plot(accel_data['Timestamp'], accel_data['X'], label='X')
plt.plot(accel_data['Timestamp'], accel_data['Y'], label='Y')
plt.plot(accel_data['Timestamp'], accel_data['Z'], label='Z')
plt.xlabel('Time')
plt.ylabel('Acceleration (g)')
plt.title('Accelerometer Data')
plt.legend()
plt.show()
```
