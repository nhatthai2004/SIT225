```python
import serial 
import time 
import csv

# Initialize serial connection
ser = serial.Serial('/dev/cu.usbmodem1101', 9600, timeout=1)  

# Define CSV name
csv_file = "DHT22_data.csv"

# Setting the CSV file
with open(csv_file, 'a', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Timestamp", "Temperature", "Humidity"])
    print("Starting data collection")
    
    timestamp = ""
    temperature = ""
    humidity = ""
    
    try:
        while True:
            # Read data from sensor through the serial 
            line = ser.readline().decode('utf-8').strip()
            
            if line:
                if line == '--onTemperatureChange':
                    # New data set starting, reset variables
                    timestamp = time.strftime('%Y%m%d%H%M%S')
                    temperature = ""
                    humidity = ""
                elif line.startswith('Temperature:'):
                    temperature = line.split(': ')[1]
                elif line.startswith('Humidity:'):
                    humidity = line.split(': ')[1]
                
                # If we have all the data, write to CSV and print
                if timestamp and temperature and humidity:
                    writer.writerow([timestamp, temperature, humidity])
                    file.flush()  # Ensure data is written to file immediately
                    print(f"{timestamp}: Temperature: {temperature}, Humidity: {humidity}")
                    
    except KeyboardInterrupt:
        # Stop data collection on keyboard interrupt
        print("Data collection stopped.")
```
