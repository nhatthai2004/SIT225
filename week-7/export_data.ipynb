```python
import serial
import csv
import time

# Set up the serial connection (replace 'COM3' with your Arduino's port)
ser = serial.Serial('/dev/cu.usbmodem2', 9600, timeout=1)
time.sleep(2)  # Allow time for the serial connection to initialize

# Open the CSV file for writing
with open('dht22.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Timestamp", "Temperature (°C)", "Humidity (%)"])

    # Record the start time
    start_time = time.time()
    duration = 30 * 60  # 30 minutes in seconds

    while True:
        # Check if 30 minutes have passed
        elapsed_time = time.time() - start_time
        if elapsed_time > duration:
            print("30 minutes have passed. Stopping data logging.")
            break

        # Read a line from the serial output
        line = ser.readline().decode('utf-8').strip()

        if line:
            # Print the line to the console (optional)
            print(line)

            # Parse the temperature and humidity from the line
            if "Temperature" in line and "Humidity" in line:
                # Split the line into components
                parts = line.split()

                # Extract temperature and humidity values
                temperature = parts[1]
                humidity = parts[4]

                # Get the current timestamp
                timestamp = time.strftime("%Y-%m-%d %H:%M:%S")

                # Write the data to the CSV file
                writer.writerow([timestamp, temperature, humidity])

        time.sleep(2)  # Match the delay in the Arduino code

# Close the serial connection when done
ser.close()
```
